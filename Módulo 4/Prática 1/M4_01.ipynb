**1) Exercício de herança múltipla e polimorfismo**
* Criar a classe **Veiculos** com um construtor que inclua marca e modelo
* Esta classe **Veículos** também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer gasolina”
* Criar a classe **VEletricos** com um construtor que inclua marca, modelo e autonomia
* Esta classe VEletricos também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer eletricidade”
* Criar a classe **BicicletaEletrica** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **VEletricos** (já que é um veículo mais elétrico que normal)
* Criar a classe **Quadriciclo** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **Veículos** (já que apenas usa a eletricidade de modo pontual)
* Criar um objeto de **BicicletaEletrica** e outro de **Quadriciclo**
* Empregar as técnicas de polimorfismo aprendidas para conseguir que ao perguntar-lhe aos objetos que acabamos de criar por reabastecimento, ambos nos respondam adequadamente


```python
# CRIAÇÃO DA CLASSE VEÍCULOS

class Veiculos():
    
    # COSTRUTOR DA CLASSE VEÍCULOS QUE INCLUI A MARCA E O MODELO 
    
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        
    # DEFINIÇÃO DO MÉTODO REABASTECER
    
    def reabastecer(self):
        return 'Este veículo tem de reabastecer gasolina.'
    
    
# CRIAÇÃO DA CLASSE VELÉTRICOS

class VEletricos():
    
    # COSTRUTOR DA CLASSE VELÉTRICOS QUE INCLUI A MARCA, O MODELO E A AUTONOMIA
    
    def __init__(self, marca, modelo, autonomia):
        self.marca = marca
        self.modelo = modelo
        self.autonomia = autonomia
        
    # DEFINIÇÃO DO MÉTODO REABASTECER
    
    def reabastecer(self):
        return 'Este veículo tem de reabastecer eletricidade.'
        
        
# CRIAÇÃO DA CLASSE BICICLETAELÉTRICA, QUE HERDA DE VEÍCULOS E DE VELÉTRICOS, DANDO PRIORIDADE A VELÉTRICOS

class BicicletaEletrica(VEletricos, Veiculos):
    pass


# CRIAÇÃO DA CLASSE QUADRICÍCLO, QUE HERDA DE VEÍCULOS E DE VELÉTRICOS, DANDO PRIORIDADE A VEÍCULOS

class Quadriciclo(Veiculos, VEletricos):
    pass


# CRIAÇÃO DE UM OBJETO DA CLASSE BICICLETAELÉTRICA E OUTRO DA CLASSE QUADRICÍCLO

bicicletaE1 = BicicletaEletrica('Peugeot Cycles', 'eT01 CrossOver', '115km')

quadriciclo1 = Quadriciclo('CFMOTO', 'CFORCE 520')

# USO DO MÉTODO REABASTECER NOS OBJETOS CRIADOS

print(bicicletaE1.reabastecer())

print(quadriciclo1.reabastecer())
```

**2) Exercício de herança**
* Criar a superclasse **Pessoal_Universitario** com um único atributo que seja um dicionário que inclua (Keys): “id”, “nome” e “email” 
* Criar as classes **Escritorio**, **Professor** e **Aluno**. Todas herdaram de **Pessoal_Universitário** e os seus dados
* No caso de **Escritorio**, deve adicionar ao seu dicionário de dados o dado “Gabinete”
* No caso de **Professor**, deve adicionar ao seu dicionário de dados o dado “Especializacao”
* No caso de **Aluno**, deve adicionar ao seu dicionário de dados o dado “CreditosAprovados" (integer)  
* Criar métodos que sirvam para mostrar a informação de cada classe
* Criar objetos de todas as classes
* Mostrar a informação desses objetos e verificar o seu output.

```python
# CRIAÇÃO DA SUPERCLASSE PESSOAL UNIVERSITÁRIO COM O ATRIBUTO DICIONÁRIO, QUE INCLUI ID, NOME E EMAIL

class Pessoal_Universitario():
    
    dic = {}
    
    def __init__ (self, ID, nome, email):
        self.dic['ID'] = ID
        self.dic['nome'] = nome
        self.dic['email'] = email
        
    # MÉTODO QUE MOSTRA A INFORMAÇÃO DO PESSOAL UNIVERSITÁRIO
    
    def __str__(self):
        return f"ID: {self.dic['ID']} - Nome: {self.dic['nome']} - Email: {self.dic['email']}"
        

# CRIAÇÃO DA CLASSE ESCRITÓRIO

class Escritorio(Pessoal_Universitario):
    
    def __init__(self,ID, nome, email, gabinete):
        super().__init__(ID, nome, email)
        self.dic['gabinete'] = gabinete
        
    # MÉTODO QUE MOSTRA A INFORMAÇÃO DO ESCRITÓRIO
    
    def __str__(self):
        return f"ID: {self.dic['ID']} - Nome: {self.dic['nome']} - Email: {self.dic['email']} - Gabinete: {self.dic['gabinete']}"
    
        
# CRIAÇÃO DA CLASSE PROFESSOR

class Professor(Pessoal_Universitario):
    
    def __init__(self, ID, nome, email, especializacao):
        super().__init__(ID, nome, email)
        self.dic['especializacao'] = especializacao
        
    # MÉTODO QUE MOSTRA A INFORMAÇÃO DOS PROFESSORES
    
    def __str__(self):
        return f"ID: {self.dic['ID']} - Nome: {self.dic['nome']} - Email: {self.dic['email']} - Especialização: {self.dic['especializacao']}"
    

# CRIAÇÃO DA CLASSE ALUNO

class Aluno (Pessoal_Universitario):
    
    def __init__(self, ID, nome, email, CreditosAprovados):
        super().__init__(ID, nome, email)
        self.dic['CreditosAprovados'] = CreditosAprovados
        
    # MÉTODO QUE MOSTRA A INFORMAÇÃO DOS PROFESSORES
    
    def __str__(self):
        return f"ID: {self.dic['ID']} - Nome: {self.dic['nome']} - Email: {self.dic['email']} - Créditos Aprovados: {self.dic['CreditosAprovados']}"

    
# CRIAÇÃO DE OBJETOS DE TODAS AS CLASSES

pessoalUniversitario1 = Pessoal_Universitario('123456789', 'João Silva', 'joaosil@gmail.com')

print(pessoalUniversitario1)

pessoalUniversitario2 = Pessoal_Universitario('987654321', 'Joana Salvador', 'jooanasal123@hotmail.com')

print(pessoalUniversitario2)

escritorio1 = Escritorio('321654987', 'Inácio Torres', 'inaciot96@gmail.com', 'A12')

print(escritorio1)

escritorio2 = Escritorio('789456123', 'Inês Lourenço', 'ilourenco33@gmail.com', 'G27')

print(escritorio2)

professor1 = Professor('192837465', 'Pedro Chaves', 'pedrochaves75@outlook.pt', 'Literatura')

print(professor1)

professor2 = Professor('918273645', 'Petra Lima', 'petralima86@outlook.pt', 'História')

print(professor2)

aluno1 = Aluno('123987456', 'Paulo Leitão', 'paleit123@gmail.com', 7)

print(aluno1)

aluno2 = Aluno('789123654', 'Ana Graça', 'anagra00@gmail.com', 6)

print(aluno2)
```
