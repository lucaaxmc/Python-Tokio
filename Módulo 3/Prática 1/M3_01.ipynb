No exercício anterior, ao fazer um print as notas e nomes, vamos obter a nota e nome do aluno. Como pode ser considerado um problema de segurança, vamos encapsular o código para que não se possa aceder diretamente às variáveis e o acesso tenha de ser através de métodos.

### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./eixos.png"  />



### Pontos e coordenadas

O objetivo é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Puntos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vector.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que significa:

5 à direita e 2 para cima.

````python
# IMPORTAÇÃO DO MÓDULO MATH PARA FAZER A RAÍZ QUADRADA

import math
math.sqrt(9)

# CRIAÇÃO DA CLASSE PONTO

class Ponto:
    __coordenadas = ''
    
    # CRIAÇÃO DO MÉTODO CONSTRUTOR DA CLASSE
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
        #DEFINIR O VALOR DEFAULT COMO 0
        
        if self.x == '':
            self.x = 0
        else: pass
        
        if self.y == '':
            self.y = 0
        else: pass
        
        # GARANTIR QUE O VALOR DAS COORDENADAS É UM NÚMERO INTEIRO
        
        self.x = int(self.x)
        self.y = int(self.y)
        
        self.__coordenadas = x, y
        
    # GETTERS E SETTERS PARA AS COORDENADAS
        
    @property
    def coordenadas(self):
        return self.__coordenadas

    @coordenadas.setter
    def coordenadas (self, novaCoordenada):
        self.__coordenadas = novaCoordenada
        
    # MÉTODO PARA CRIAR PONTOS FACILMENTE
        
    def criar_ponto(self):
        
        self.x, self.y = [i for i in input('Introduza ponto final X e Y separados por uma virgula:').split(',')]
        if self.x == '':self.x = 0
        if self.y == '':self.y = 0
        self.x = int(self.x)
        self.y = int(self.y)
        self.coordenadas = self.x, self.y
        return f'Ponto1: {self.x},{self.y}'
        
    # DEFINIÇÃO DO MÉTODO STR PARA IMPRIMIR AS COORDENADAS NO FORMATO (X,Y)
    
    def __str__(self):
        return '({}, {})'.format(self.x,self.y)
    
    # CRIAÇÃO DO MÉTODO QUADRANTE 
    
    def quadrante(self):
        
        if self.x == 0 and self.y == 0:
            return 'Origem'
            
        elif self.x > 0 and self.y > 0:
            return '1º Quadrante'
            
        elif self.x < 0 and self.y > 0 :
            return '2º Quadrante'
            
        elif self.x < 0 and self.y < 0:
            return '3º Quadrante'
            
        elif self.x > 0  and self.y < 0:
            return '4º Quadrante'   
            
    # CRIAÇÃO DO MÉTODO VETOR
    
    def vetor(self, ponto2):
        x2 = ponto2[0]
        y2 = ponto2[1]
        
        AB = (self.x-x2,self.y-y2)
        
        return f'{AB}'

    # CRIAÇÃO DO MÉTODO DISTÂCIA
    
    def distancia(self, ponto3):
        x3 = ponto3[0]
        y3 = ponto3[1]

        distancia = (x3-self.x)**2 + (y3-self.y)**2

        return '{:.02f}'.format(math.sqrt(distancia))
    

    
    
# CRIAÇÃO DA CLASSE RETÂNGULO

class Retangulo(Ponto):
    
    # CRIAÇÃO DO MÉTODO CONSTRUTOR DA CLASSE
    
    def __init__(self, inicial = [], final = []):
        self.x1 = inicial[0]
        self.y1 = inicial[1]
        
        self.x2 = final[0]
        self.y2 = final[1]
        
        self.inicial = inicial
        self.final = final
        
    # MÉTODO PARA CRIAR PONTOS FACILMENTE
    
    def inserir_pontos(self):
        self.x1,self.y1 =[i for i in input('Introduza ponto final X e Y separados por uma virgula:').split(',')]
        self.x2,self.y2 =[i for i in input('Introduza ponto final X e Y separados por uma virgula:').split(',')]
        
        #DEFINIR O VALOR DEFAULT COMO 0
        
        if self.x1 == '':
            self.x1 = 0
        if self.y1 == '':
            self.y1 = 0
        if self.x2 == '':
            self.x2 = 0
        if self.y2 == '':
            self.y2 = 0
            
        self.inicial[0] = int(self.x1)
        self.inicial[1] = int(self.y1)
        
        self.final[0] = int(self.x2)
        self.final[1] = int(self.y2)
        
        return f'ponto1: {self.inicial[0]},{self.inicial[1]} - ponto2: {self.x2},{self.y2}'
    
    # DEFINIÇÃO DO MÉTODO STR PARA IMPRIMIR AS COORDENADAS
    
    def __str__(self):
        return f'{self.inicial} - {self.final}'
    
    # DEFINIÇÃO DO MÉTODO BASE, PARA CALCULAR E MOSTRAR A MEDIDA DA BASE DO RETÂNGULO
    
    def base(self):
        base = abs(self.x1 - self.x2)
        return f'O retangulo tem {base} unidades de base'
    
    # DEFINIÇÃO DO MÉTODO ALTURA, PARA CALCULAR E MOSTRAR A MEDIDA DA ALTURA DO RETÂNGULO
    
    def altura(self):
        altura = abs(self.y1 - self.y2)
        return f'O retangulo tem {altura} unidades de altura'
    
    # DEFINIÇÃO DO MÉTODO ÁREA, PARA CALCULAR E MOSTRAR A ÁREA DO RETÂNGULO
    
    def area(self):
        area = (self.x1-self.x2) * (self.y1-self.y2)
        return f'O retangulo tem {abs(area)} u.a de area'
    
    
    

# EXPERIMENTAÇÃO

# CRIAÇÃO DOS 4 PONTOS PEDIDOS

A = Ponto(2, 3)
B = Ponto(5, 5)
C = Ponto(-3, -1)
D = Ponto(0, 0)

# IMPRESSÃO DOS PONTOS

print('A:', str(A))
print('B:', str(B))
print('C:', str(C))
print('D:', str(D))
print()

# VERIFICAÇÃO DO QUADRANTE DE CADA PONTO

print('A -', A.quadrante())
print('B -', B.quadrante())
print('C -', C.quadrante())
print('D -', D.quadrante())
print()

# CÁLCULO DOS VETORES AB E BA
print('Vetor AB -', A.vetor(B.coordenadas))
print('Vetor BA -', B.vetor(A.coordenadas))
print()

# CÁLCULO DA DISTÂNCIA ENTRE 'A e B' E 'B e A'

print('A distancia de A para B é de :', A.distancia(B.coordenadas))
print('A distancia de B para A é de :', B.distancia(A.coordenadas))
print()

# DETERMINAR QUAL DOS PONTOS SE ENCONTRA MAIS LONGE DA ORIGEM
if B.distancia(D.coordenadas) > A.distancia(D.coordenadas) and B.distancia(D.coordenadas) > C.distancia(D.coordenadas):
        print('O ponto mais distante da origem é o ponto  B.')
        
if A.distancia(D.coordenadas) > B.distancia(D.coordenadas) and A.distancia(D.coordenadas) > C.distancia(D.coordenadas):
        print('O ponto mais distante da origem é o ponto  A.')
        
if C.distancia(D.coordenadas) > B.distancia(D.coordenadas) and C.distancia(D.coordenadas) > A.distancia(D.coordenadas):
        print('O ponto mais distante da origem é o ponto  C.')
        
print()
        
# DEFINIR UM RETÂNGULO COM OS PONTOS A E B

retangulo1 = Retangulo(A.coordenadas,B.coordenadas)

#CALCULAR A BASE, ALTURA E ÁREA DO RETÂNGULO

print(retangulo1.base())
print(retangulo1.altura())
print(retangulo1.area())
````
